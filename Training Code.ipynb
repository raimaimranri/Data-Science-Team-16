```python
import os
from dotenv import load_dotenv
import psycopg2
from vanna.remote import VannaDefault

# Load environment variables from .env file
load_dotenv()

# Load API key from environment variables
api_key = os.getenv("api_key")

# Define the Vanna model name
vanna_model_name = 'vanna-ai'

# Initialize Vanna
vn = VannaDefault(model=vanna_model_name, api_key=api_key)

# Connect to PostgreSQL database
conn = psycopg2.connect(
    host=os.getenv('localhost'),
    dbname=os.getenv('ds2'),
    user=os.getenv('postgres'),
    password=os.getenv('123456')
)
cursor = conn.cursor()

# Function to query the database using natural language
def query_db(natural_language_query):
    # Generate SQL query using Vanna AI
    sql_query = vn.generate_sql(natural_language_query)
    
    # Execute the SQL query
    cursor.execute(sql_query)
    
    # Fetch all results
    results = cursor.fetchall()
    return results

# Example usage
results = query_db("Show me the biggest table names")
for row in results:
    print(row)

# Close the database connection
cursor.close()
conn.close()
```
